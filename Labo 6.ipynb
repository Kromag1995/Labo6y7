```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import os
```

```python
%matplotlib notebook
```

```python
def binarizacion(picos, mom):
    if mom == "median":
        filtro = np.median(maximos_filtrados_y)
    elif mom == "mean":
        filtro = np.mean(maximos_filtrados_y)
    serie_binaria = []
    for i in picos:
        if i > filtro:
            serie_binaria.append(1)
        else:
            serie_binaria.append(0)
    return serie_binaria, filtro
```

```python
carpeta_datos = open("./Muestras sin saturacion/DATA_20190307-0003_4_10Millones.bin", "rb")
a = np.fromfile(carpeta_datos, dtype=np.float)
carpeta_datos.close()
```

```python
plt.plot(a[:])
plt.grid()
```

```python
filtrados_x = []
filtrados_y = []
for index,dato in enumerate(a):
    if dato == np.inf or dato<0.001:
        pass
    else:
        filtrados_x.append(index)
        filtrados_y.append(dato)
```

```python
i_p = 0
f_p = -1
indexs =signal.find_peaks(filtrados_y[i_p:f_p])[0]
maximos_filtrados_y = []
maximos_filtrados_x = []
for i in indexs:
    maximos_filtrados_y.append(filtrados_y[i])
    maximos_filtrados_x.append(filtrados_x[i])
```

```python
plt.plot(filtrados_x[:2000],filtrados_y[:2000], '.-')
plt.plot(maximos_filtrados_x[:20],maximos_filtrados_y[:20], 'or')
plt.grid()
```

```python
indexs =signal.find_peaks(filtrados_y, height=0.05,distance=15)[0]
```

```python
maximos_filtrados_y = []
maximos_filtrados_x = []
for i in indexs:
    maximos_filtrados_y.append(filtrados_y[i])
    maximos_filtrados_x.append(filtrados_x[i])
```

```python
i_f, f_f, i_m, f_m = 0,20000, 0,90
plt.plot(filtrados_x[i_f:f_f],filtrados_y[i_f:f_f], '.-')
plt.plot(maximos_filtrados_x[i_m:f_m],maximos_filtrados_y[i_m:f_m], 'or')
```

```python
fig, ax = plt.subplots()
ax.stem(maximos_filtrados_x[:100],maximos_filtrados_y[:100])
plt.grid()
plt.show()
```

```python
dif_temporal = []
for i in range(len(maximos_filtrados_x)-1):
    dif_temporal.append(maximos_filtrados_x[i+1]-maximos_filtrados_x[i])
```

```python
fig, ax = plt.subplots()
i,f = 450,570
ax.stem(maximos_filtrados_x[i:f],dif_temporal[i:f])
plt.grid()
```

```python
binario_mean,mean = binarizacion(maximos_filtrados_y, "mean")
binario_median,median = binarizacion(maximos_filtrados_y, "median")
print(mean, median, mean-median)
```
