```python
#abro los archivos de la carpeta donde esten los datos
datos=[]
with os.scandir('/home/dell/Documents/0.339223719241889Labo 6/data/Muestras sin saturacion/') as todosdatos:
    for archivo in todosdatos:
        print(archivo.name)
        if archivo.name != 'Readme.txt':
            datos.append(open('/home/dell/Documents/Labo 6/data/Muestras sin saturacion/'+archivo.name))
```

```python
#convierto los datos a float
tira = []
for i in range(len(datos)):
    tira.append(np.fromfile(datos[i],dtype=float))
```

```python
tiempos = {}
salto=102.4*(1e-9)
for j in range(len(tira)):
    tiempos[f't_{j}']=np.arange(0,salto*len(tira[j]),salto)
```

```python
#importante chequear que esos dos numeros sean iguales! si no lo son es porque quedó mal definido el tiempo
print(len(tiempos['t_4']),len(tira[4]))
```

```python
#filtramos los datos, sacado del ruido
filtrado_V={}
filtrado_t={}
for j in range(len(tira)):  
    filtrado_V[f'fil_V_{j}']=[]
    filtrado_t[f'fil_t_{j}']=[]
    for i in range(len(tira[j])):
        if tira[j][i] > 0.001:
        #if tira_1[i] == np.inf:
         #   pass
        #else:
            filtrado_V[f'fil_V_{j}'].append(tira[j][i])
            filtrado_t[f'fil_t_{j}'].append(tiempos[f't_{j}'][i])
```

```python
#máximos
maximos_V={}
maximos_t={}
for j in range(len(tira)):  
    maximos_V[f'max_V_{j}']=[]
    maximos_t[f'max_t_{j}']=[]
    peaks, _ = find_peaks(tira[j],height=[1],distance = 150)
    for i in peaks:
        maximos_V[f'max_V_{j}'].append(tira[j][i])
        maximos_t[f'max_t_{j}'].append(tiempos[f't_{j}'][i])
```
